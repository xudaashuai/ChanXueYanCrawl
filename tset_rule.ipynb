# 这个是测试规则的脚本

## 这里包含需要用到的库

```python
import re # 正则
from xtp import * # 我写的 用于规则匹配的 Word和Sentence类
import pymongo # mongo数据库连接
```

## 连接数据库

```python
connection = pymongo.MongoClient('localhost', 27017)
tdb = connection.data
comments = tdb.comments
```

## 定义两种测试方法

```python
def test_by_re(test_num,page,rule):
    num = 0
    for i in range(test_num//page):
        for k,item in enumerate(comments.find({'hanlp_result':{'$ne':None}}).limit(page).skip(i * page)):
            hanlp_result = item['hanlp_result']
            s = Sentence(hanlp_result)
            result = rule.findall(s.str())
            num += len(result)!=0
            print(num,'/',i*page+k,'/',test_num,*result)
```

```python
def test_by_dict(test_num,page,ms):
    num = 0
    for i in range(test_num//page):
        for k,item in enumerate(comments.find({'hanlp_result':{'$ne':None}}).limit(page).skip(i * page)):
            hanlp_result = item['hanlp_result']
            s = Sentence(hanlp_result)
            result = s.match(ms)
            num += len(result)!=0
            print(num,'/',i*page+k,'/',test_num,*result)
```

## 使用正则方法匹配规则

```python
s = '([^ ^#]+) n.? \d+ 主谓关系(?:#([^ ]+) d.? \d+ [^#]+)*#([^ ]+) a.? [^ ]+ [^#]+'
r = re.compile(s)
test_by_re(10000,100,r)
```

## 使用简单字典方法匹配规则

```python
ms = [[
    {
        'POSTAG': 'n.?',
        'HEAD': '1',
        'DEPREL':'主谓关系'
    }, {
        'POSTAG': 'a.?'
    }
],[
    {
        'POSTAG': 'n.?',
        'HEAD': '2',
        'DEPREL':'主谓关系'
    }, {
        'POSTAG': 'd.?'
    }, {
        'POSTAG': 'a.?'
    }
],[
{
        'POSTAG': 'n.?',
        'HEAD': '3',
        'DEPREL':'主谓关系'
    }, {
        'POSTAG': 'd.?'
    }, {
        'POSTAG': 'd.?'
    }, {
        'POSTAG': 'a.?'
    }
]]
test_by_dict(100,10,ms)
```
